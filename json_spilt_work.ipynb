```python
import os
import shutil
import json
```

```python
path = '/content/gdrive/MyDrive/Colab Notebooks/IT401_Project/Data'
```

```python
dir = path + '/Dataset'
for files in os.listdir(dir):
    path1 = os.path.join(dir, files)
    try:
        shutil.rmtree(path1)
    except OSError:
        os.remove(path1)
```

```python
dir_list = os.listdir(path + '/Dataset')
len(dir_list)
```

```python
path_test_old = "/content/gdrive/MyDrive/Colab Notebooks/IT401_Project/Data/testing_data/video"
path_train_old = "/content/gdrive/MyDrive/Colab Notebooks/IT401_Project/Data/training_data/video"
path_test_new = "/content/gdrive/MyDrive/Colab Notebooks/IT401_Project/Data/test_new"
path_train_new = "/content/gdrive/MyDrive/Colab Notebooks/IT401_Project/Data/train_new"
```

```python
for video in os.listdir(path_test_old):
  shutil.copy2(os.path.join(path_test_old,video), path+'/Dataset')


for video in os.listdir(path_train_old):
  shutil.copy2(os.path.join(path_train_old,video), path+'/Dataset')
```
